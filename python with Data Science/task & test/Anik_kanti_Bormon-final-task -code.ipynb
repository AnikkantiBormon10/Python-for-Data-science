#2 Ans : Concatenate NumPy Arrays
For combining a large number of DataFrames, it is generally better to use the concat() method rather than append().

Reasons to Choose concat() for Large Numbers of DataFrames
Performance:

concat() is more efficient when combining many DataFrames, as it’s optimized for batch concatenation.
append() is essentially a wrapper around concat() that is better suited for adding a small number of DataFrames or rows at a time, rather than a large collection.
Ease of Use:

concat() can take a list of DataFrames and combine them in one go, which is ideal for merging many DataFrames at once.
With append(), you would need to use a loop or chain multiple append() calls, which is less efficient and can make the code harder to read.

**Why Use groupby() for Analyzing Datasets with Multiple Categories?**


*Using groupby() is valuable when analyzing datasets with multiple categories because it allows you to:*


**Summarize Data**: Quickly calculate aggregated metrics (e.g., sum, average) for
each category, providing insights into overall patterns and trends.


**Segment Data**: Understand how different categories (e.g., regions, products) contribute to results, making it easier to identify which segments are performing well.


**Optimize Performance**: Perform calculations in groups without having to filter data repeatedly, which is efficient for large datasets.

**Data Loading and Handling Missing Values**

* Load the Iris dataset from Scikit-Learn.

* Simulate missing values: Introduce a few missing values randomly into the dataset. For example, set 5 random entries in the sepal length and petal length columns to NaN.

* Handle missing data: Use the mean of each column to fill in these missing values.


**Data Splitting and K-Fold Cross-Validation**

* Split the dataset into features (X) and target (y).

* Implement K-Fold Cross-Validation with 5 folds using a Decision Tree Classifier.

* Calculate evaluation metrics (accuracy, precision, recall, and F1 score) across all folds.


**Hyperparameter Tuning**
* Tune hyperparameters (max_depth, min_samples_split) using GridSearchCV with Cross-Validation.
* Identify the best hyperparameters and save them for later use.

[Same example but with KNN, SVC]


````python
#1 Data Loading and Handling Missing Values

from sklearn.datasets import load_iris
````

````python
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
````

````python
np.random.seed(0)
missing_indices_sepal = np.random.choice(df.index, 5, replace=False)
missing_indices_petal = np.random.choice(df.index, 5, replace=False)
df.loc[missing_indices_sepal, 'sepal length (cm)'] = np.nan
df.loc[missing_indices_petal, 'petal length (cm)'] = np.nan
````

````python
df.fillna(df.mean(), inplace=True)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
````

````python
X = iris.data
y = iris.target

model = DecisionTreeClassifier(random_state=1)

kf = KFold(n_splits=5, shuffle=True, random_state=1)

cv_scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')

cv_scores
````

````python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=42)

model = KNeighborsClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy: ", accuracy_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred, average='macro'))
print("Recall: ", recall_score(y_test, y_pred, average='macro'))
print("F1 score: ", f1_score(y_test, y_pred, average='macro'))

cross_val_scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
print("Cross validation accuracy scores: ", cross_val_scores)
````

````python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

# Define the hyperparameter grids for each model

# Decision Tree hyperparameter grid
param_grid_dt = {
    'max_depth': [3, 5, 7, None],
    'min_samples_split': [2, 5, 10]
}

# Perform GridSearchCV for each model
# Decision Tree
grid_search_dt = GridSearchCV(DecisionTreeClassifier(random_state=1), param_grid_dt, cv=5, scoring='accuracy')
grid_search_dt.fit(X, y)

grid_search_dt.best_score_

````

````python
# KNN hyperparameter grid
param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}


# K-Nearest Neighbors
grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5, scoring='accuracy')
grid_search_knn.fit(X, y)
grid_search_knn.best_score_

````

````python

# SVC hyperparameter grid
param_grid_svc = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

# Support Vector Classifier
grid_search_svc = GridSearchCV(SVC(random_state=1), param_grid_svc, cv=5, scoring='accuracy')
grid_search_svc.fit(X, y)
grid_search_svc.best_score_

````

````python
# Store best hyperparameters for each model
best_params_dt = grid_search_dt.best_params_
best_params_knn = grid_search_knn.best_params_
best_params_svc = grid_search_svc.best_params_

best_params_dt, best_params_knn, best_params_svc
````

````python
# Sales data
categories = ['Fruits', 'Vegetables', 'Dairy', 'Bakery', 'Snacks']
units_sold = [120, 200, 150, 90, 130]
avg_price = [2.5, 1.5, 2.0, 1.8, 3.0]  # in dollars
customer_ratings = [4.5, 4.2, 4.0, 3.8, 4.1]

# convenience
data = {
    'Categories': categories,
    'Units Sold': units_sold,
    'Avg Price': avg_price,
    'Customer Ratings': customer_ratings
}
df = pd.DataFrame(data)

# 1. Bar-Plot
plt.figure(figsize=(9, 5))
plt.bar(df['Categories'], df['Units Sold'], color='skyblue')
plt.title('Units Sold by Category')
plt.xlabel('Categories')
plt.ylabel('Units Sold')
plt.xticks(rotation=45)
plt.legend(['Units Sold'])
plt.tight_layout()
plt.show()

````

````python
# 2. Scatter Plot
plt.figure(figsize=(7, 4.5))
plt.scatter(df['Units Sold'], df['Avg Price'], color='green')
plt.title('Units Sold vs. Average Price')
plt.xlabel('Units Sold')
plt.ylabel('Average Price ($)')
plt.grid(True)
plt.tight_layout()
plt.show()

````

````python
# 3. Histogram
plt.figure(figsize=(5, 3))
plt.hist(df['Customer Ratings'], bins=5, color='orange', edgecolor='black')
plt.title('Distribution of Customer Ratings')
plt.xlabel('Customer Ratings')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

````

````python
# 4. Pie Chart
import seaborn as sns


plt.figure(figsize=(6, 6))
plt.pie(df['Units Sold'], labels=df['Categories'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title('Percentage of Units Sold by Category')
plt.axis('equal')
plt.tight_layout()
plt.show()

````

````python
# 5. Box Plot
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['Customer Ratings'], color='black')
plt.title('Box Plot of Customer Ratings')
plt.ylabel('Customer Ratings')
plt.grid(axis='y')
plt.tight_layout()
plt.show()


````
