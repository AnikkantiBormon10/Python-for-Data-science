```python
import matplotlib.pyplot as plt
```

# Bar plot

```python
x=[1,2,3,4]
y=[4,5,6,7]
c=['red','green','blue','yellow']
plt.scatter(x, y, c=c, alpha=.1)
plt.show()
# plt.bar(x,y,color=c)
# plt.show()
```

```python
x=['python','java','c','c++']
y=[85,60,70,95]
plt.xlabel("Languages",fontsize=14)
plt.ylabel("No of Studnts.")
plt.title("Popularity of Programming Language")
plt.bar(x,y,width=0.4,color='blue',edgecolor='g',linewidth=5,linestyle="-",alpha=0.5,label="Matplotlib")
plt.legend()
plt.show()
```

```python
x=['python','java','c','c++']
y=[85,60,70,95]
z=[30,56,88,79]
plt.xlabel("Languages",fontsize=14)
plt.ylabel("No of Studnts.")
plt.title("Popularity of Programming Language")
plt.bar(x,y,width=0.4,color='blue',edgecolor='r',linewidth=5,linestyle=":",alpha=0.5,label="Matplotlib")
plt.bar(x,z,width=0.4,color='black',edgecolor='r',linewidth=5,linestyle=":",alpha=0.4,label="Matplotlib1")
plt.legend()
plt.show()
```

# Scatter Plot

```python
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
size=[30,40,50,60,70,30,50]
c=['r','g','b','orange','y','g','b']
plt.scatter(day,no,s=size,c=c,alpha=0.9,marker='p',label="Scatter Plot")
plt.legend()
plt.colorbar().set_label('ColorBar')
plt.title('Scatter Plot')
plt.xlabel('day',fontsize=20)
plt.ylabel('no.',fontsize=18)
plt.show()
```

# Histogram

```python
import numpy as np
import random
x=np.random.randint(1,100,10)
l=[15,23,40,50,60]
plt.hist(x,color='g',bins=l,edgecolor='r')
# plt.hist?
plt.show()
```

# Pie Chart

```python
program=['python','java','c','c++']
no_of_stud=[85,60,70,95]
plt.pie(no_of_stud)
# plt.show()
c=['orange','gray','blue','yellow']
plt.pie(no_of_stud,labels=program,colors=c,autopct='%1.2f%%',shadow=True)
plt.show()
```

# Stem Plot

```python
x=[1,2,3,4,5]
y=[2,2,4,5,6]
plt.stem(x,y,linefmt='--',label='Stem Plot')
plt.legend()
plt.show()
```

# Box  Plot

```python
x=[20,30,40,50,60,70,80]
plt.boxplot(x,labels=['Box Plot'],vert=False)
plt.legend()
plt.show()
```
